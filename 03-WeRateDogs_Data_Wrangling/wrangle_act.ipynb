### Define (3) Combine 'doggo', 'floofer', etc. into 'descriptions' column

Make new 'description' column combining four descriptions, drop original four. Make new string from all four values. remove 'nan' substrings, drop individual columns.

### Code:

````python
# make new column
descriptions = []
for a,b,c,d in zip(df_clean_eta['doggo'],df_clean_eta['floofer'],\
                   df_clean_eta['pupper'],df_clean_eta['puppo']):
        
    description = str(a)+'-'+str(b)+'-'+str(c)+'-'+str(d)
    if description=='nan-nan-nan-nan': description = 'none'
    else: 
        description = description.replace('-nan','')
        description = description.replace('nan-','')
    descriptions.append(description)

# add new column
df_clean_eta['description'] = descriptions

# remove old columns
df_clean_eta = df_clean_eta.drop(['doggo','floofer','pupper','puppo'], axis=1)
````

Chose the best prediction from confidence levels and whether predictions are of dogs. Iterate through p? and p?_dog, to save highest prediction of a breed. Save 'none' string if no prediction of a dog is made. Drop individual columns.

## 2) Breed Distribution Ratings

````python
breed_ratings = twitter_archive_master['rating_%']
no_breed_rating = breed_ratings[twitter_archive_master.breed=='none']
rare_breed_rating = {breed:breed_ratings[twitter_archive_master.breed==breed].values for breed in rare_breeds.index}
uncommon_breed_rating = {breed:breed_ratings[twitter_archive_master.breed==breed].values for breed in uncommon_breeds.index}
common_breed_rating = {breed:breed_ratings[twitter_archive_master.breed==breed].values for breed in common_breeds.index}
````

````python
# Following https://stackoverflow.com/questions/8671808/avoiding-overlapping-datapoints-in-a-scatter-dot-beeswarm-plot
def rand_jitter(arr):
    stdev = 0.01 * (max(arr) - min(arr))
    return arr + np.random.randn(len(arr)) * stdev

def jitter(x, y):
    return plt.scatter(rand_jitter(x), rand_jitter(y), s=50, marker='o',alpha=0.5)
````

````python
plt.figure(figsize=(6,6))
#no breed prediction
ratings = no_breed_rating.values
counts = np.ones(len(ratings))*len(ratings)
jitter(counts,ratings)

plt.title('No Breed Prediction')
plt.ylabel('Dog Ratings')
plt.xlabel('Breed Counts')
plt.tight_layout()
plt.savefig(os.path.join(output_dir, 'no_breed_ratings.png'))
plt.show()
````

````python
plt.figure(figsize=(6,6))
#rare breeds
for breed in rare_breed_rating:
    ratings = rare_breed_rating[breed]
    counts = np.ones(len(ratings))*len(ratings)
    jitter(counts,ratings)

plt.title('Rare Breeds')
plt.ylabel('Dog Ratings')
plt.xlabel('Breed Counts')
plt.tight_layout()
plt.savefig(os.path.join(output_dir, 'rare_breed_ratings.png'))
plt.show()
````

````python
plt.figure(figsize=(6,6))
#uncommon breeds
for breed in uncommon_breed_rating:
    ratings = uncommon_breed_rating[breed]
    counts = np.ones(len(ratings))*len(ratings)
    jitter(counts,ratings)

plt.title('Uncommon Breeds')
plt.ylabel('Dog Ratings')
plt.xlabel('Breed Counts')
plt.tight_layout()
plt.savefig(os.path.join(output_dir, 'uncommon_breed_ratings.png'))
plt.show()
````

````python
plt.figure(figsize=(6,6))
#common breeds
for breed in common_breed_rating:
    ratings = common_breed_rating[breed]
    counts = np.ones(len(ratings))*len(ratings)
    jitter(counts,ratings)

plt.title('Common Breeds')
plt.ylabel('Dog Ratings')
plt.xlabel('Breed Counts')
plt.tight_layout()
plt.savefig(os.path.join(output_dir, 'common_breed_ratings.png'))
plt.show()
````

> I have plotted the spread of ratings for each of the breed groups, including 146 posts that included no prediction of the breed. Although, the Rare Breed group is inheriently undersampled the spread of ratings appear to be higher on average than the others. Additionally, the Uncommon Breeds appear to average higher than the Common breeds.
